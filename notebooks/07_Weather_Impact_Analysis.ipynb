```python
import pandas as pd 

orders = pd.read_csv("../Data/processed/clean_orders.csv")
weather = pd.read_csv("../Data/raw/weather_daily.csv")

orders['date'] = pd.to_datetime(orders['order_placed_at']).dt.date
weather['date'] = pd.to_datetime(weather['date']).dt.date

merged = pd.merge(orders, weather, on='date', how='left')

merged.head()
```

```python
merged['rain_type'] = merged['rain_sum'].apply(
    lambda x: 'Rainy' if x > 0 else 'No Rain'
)
```

```python
merged['temp_type'] = pd.cut(
    merged['temperature_2m_max'],
    bins=[0,25,35,50],
    labels=['Cool','Moderate','Hot']
)
```

```python
merged.to_csv("../Data/processed/orders_with_weather.csv", index=False)
```
