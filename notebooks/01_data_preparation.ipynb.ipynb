```python
import pandas as pd
orders = pd.read_csv("../data/raw/order_history.csv")

print("Shape:", orders.shape)
orders.head()
```

```python
orders.columns = orders.columns.str.lower().str.replace(" ", "_")
orders.columns
```

```python
orders['order_placed_at'] = pd.to_datetime(orders['order_placed_at'])
```

```python
orders.dtypes
```

```python
orders['distance_km'] = (
    orders['distance']
    .str.replace('<', '', regex=False)
    .str.replace('km', '', regex=False)
    .astype(float)
)
```

```python
orders[['items_in_order']].head()
```

```python
orders['items_in_order'].unique()[:10]
```

```python
orders['item_count'] = orders['items_in_order'].str.count(' x ')
orders[['items_in_order','item_count']].head()
```

```python
orders.to_csv("../data/processed/clean_orders.csv", index=False)
```
